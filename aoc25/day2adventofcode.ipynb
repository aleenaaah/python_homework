```python
# how to split the numbers? total length divided by 2 is the half 
def invalid_numbers(n):
    s = str(n)
    if len(s) % 2 == 1:
        return False
    half = len(s) // 2
    first = s[:half]        
    second = s[half:]      

    return first == second #means if first = second, return true, else return false 
    
```

```python
for n in range(998, 1012):
    if invalid_numbers(n):
        print("invalid", n)
```

```python
def repeatedSubstringPattern(s):
    n = len(s)
    
    # Go through divisors of n
    for i in range(1, n):
        if n % i == 0:  # Only check if i divides length
            x = s[:i]
            if x * (n // i) == s:
                return True
    return False
```
